## This script will create pdfs of both model output and TRMM data for certain gridpoints

```python
import matplotlib.pyplot as plt
import xarray as xr
import numpy as np
import pandas as pd
```

```python
trmm14=xr.open_dataset('../parameterization_exp/trmm_2014_JunAug.nc')
trmm15=xr.open_dataset('../parameterization_exp/trmm_2015_JunAug.nc')
ti,la,lo=np.shape(trmm14.PPT)
trmm=np.zeros([184,la,lo])
trmm[:92,:,:]=trmm14.PPT
trmm[92:,:,:]=trmm15.PPT
print(np.shape(trmm))
print(trmm14.lat[0].data)
```

```python
chiangsaen_loc={'lat':20.27416667,'lon':100.09083333}
mukdahan_loc={'lat':16.54000000,'lon':104.73666667}
kratie_loc={'lat':12.47666667,'lon':106.01500000}
print(chiangsaen_loc['lat'])
```

```python
def pos(loc):
    for i,val in enumerate(trmm14.lat.data):
        if loc['lat']<val:
            ilat=i-1
            break
    for j,val in enumerate(trmm14.lon.data):
        if loc['lon']<val:
            ilon=j-1
            break
    return ilat,ilon
```

```python
a,b=pos(chiangsaen_loc)
print(trmm14.lat[a],trmm14.lon[b])

def plotpdf(loc):
    ilat,ilon=pos(loc)
    val=pd.DataFrame(trmm[:,ilat,ilon])
    val.plot.density()
    val.plot.hist(ylim=[0,200],xlim=[0.01,80],bins=20)
```

```python
plotpdf(chiangsaen_loc)
plotpdf(mukdahan_loc)
plotpdf(kratie_loc)
plt.show()
```
